# Analysis of Gspacer on HWMCC20 Rules (Oct. 29)

- timeout: 900s
- branch: gspacer

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (4, 4)
```

```python
filename = 'gspacer.hwmcc20-rules.lilla-my.29_10_2020-t03-55-57.csv'
timeout = 900
df = pd.read_csv(filename)
df['status'].fillna(value='(no status)', inplace=True)
df['time'].fillna(value=timeout, inplace=True)
# show a snapshot of data for visual validation
df.head()
```

```python
# quick overview of solved/unsolved instances
df[['status', 'index']].groupby(['status']).count()
```

```python
# Find unknown results. Usually means that something went wrong, not just timeout.
df.query('status == "unknown"')[['index', 'time', 'memory', 'status']].sort_values('time')
```

```python
df.query('status == "(no status)"')[['index', 'time', 'memory', 'status']].sort_values('time')
```

## HWMCC20 modelcheckers

```python
pd.read_csv('http://fmv.jku.at/hwmcc20/hwmcc20-combined-all.txt', delim_whitespace=True)
```

```python
# The values of the result column are unknown if all model checkers
# gives unknown, or if the number of model checkers that says sat
# and the number of those that say unsat tie. Otherwise, the values
# are sat or unsat based on the majority of the model checkers.
hwmcc20_results: pd.DataFrame = pd.read_csv('../hwmcc20-results.csv')
hwmcc20_results['index'] = hwmcc20_results['index'].apply(lambda x: x + '.smt2')
hwmcc20_results.head()
```

```python
hwmcc20_results[['result', 'index']].groupby(['result']).count()
```

## Compare Gspacer and HWMCC20 modelcheckers

```python
m: pd.DataFrame = pd.merge(df, hwmcc20_results, on='index')
m.head()
```

```python
m[['status', 'index', 'result']].groupby(['status', 'result']).count()
```

```python
# {sat|uns|unk}_count: The number of model checkers
#                      that gives {sat|unsat|unknown}
m.query('status == "sat" & result == "unsat"')[
    ['index', 'status', 'result', 'time', 'memory', 'sat_count', 'uns_count',
     'unk_count']].sort_values('time')
```

For `bv-vcegar_arrays_itc99_b12_p2.smt2` and `bv-rast-p17.smt2`, only nmtip says unsat.

```python
m.query('status == "unsat" & result == "sat"')[
    ['index', 'status', 'result', 'time', 'memory', 'sat_count', 'uns_count',
     'unk_count']].sort_values('time')
```

```python
m.query('status == "sat" & result == "unknown"')[
    ['index', 'status', 'result', 'time', 'memory', 'sat_count', 'uns_count',
     'unk_count']].sort_values('time')
```

```python
m.query('status == "unsat" & result == "unknown"')[
    ['index', 'status', 'result', 'time', 'memory', 'sat_count', 'uns_count',
     'unk_count']].sort_values('time')
```

```python
def scatter_values(xs, ys, xlabel, ylabel, x_min, x_max, y_min, y_max):
    assert len(xs) == len(ys)
    
    print('No. values: {}'.format(len(xs)))
    print('Min, {:s}: {} \t {:s}: {}'.format(xlabel, min(xs), ylabel, min(ys)))
    print('Max, {:s}: {} \t {:s}: {}'.format(xlabel, max(xs), ylabel, max(ys)))
    print('Avg, {:s}: {:0.2f} \t {:s}: {:0.2f}'.format(xlabel, np.mean(xs), ylabel, np.mean(ys)))
    
    # min_val = min(0, min(xs), min(ys))
    # max_val = max(max(xs), max(ys))
    plt.xlim(x_min, x_max)
    plt.ylim(y_min, y_max)
    
    plt.scatter(xs, ys)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    
    mid_x = np.linspace(x_min, x_max, 2)
    mid_y = np.linspace(y_min, y_max, 2)
    plt.plot(mid_x, mid_y, 'r-')
    plt.title("{:s} vs. {:s}".format(ylabel, xlabel))
    plt.draw()
```

```python
scatter_values(m['time'], m['avg_real'], "Gspacer Time", "Avg HWMCC20 Real Time",
               0, 900, 0, 900)
```
